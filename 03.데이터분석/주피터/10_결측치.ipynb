```python
df3 = df.copy()
df3.fillna({'SW특기':'없음'}, inplace=True)
df3
```

```python
#학교를 모두 null로 바꾸기
import numpy as np
```

```python
df4 = df.copy()
df4['학교'] = np.nan
df4
```

```python
df4.fillna('모름', inplace=True)  #모든 null을 '모름'으로 수정
df4
```

```python
df5 = df.copy()
df5['학교'] = np.nan
df5.fillna({'학교':'코스모', 'SW특기':'없음'}, inplace=True)  #각각 널값 다르게 채우기
```

```python
df5
```

```python
df6=df.copy()
#5번의 학교 null
df6.loc['5번', '학교'] = np.nan
df6.ffill()  #앞의 값으로 채워짐
df6.bfill(inplace=True)  #뒤의 값으로 채워짐
df6
```

```python
df7 = df.copy()
df7.loc['1번', '학교'] = np.nan
df7.dropna()  #none값이 포함된 행 제거 = (axis='index')?
df7.dropna(axis='columns', how = 'any')  #none값 열 제거
df7
```

```python
df7.loc['9번'] = [np.nan] * 9  #새로운 행(9번)을 null로 채우기
df7.dropna(axis='index', how='all')  #전부 null값인 행만 없애기
```

```python
df7 = df.copy()
df7.loc['9번'] = [np.nan] * 9  #새로운 행(9번)을 모두 null로 채우기
df7['학년'] = np.nan  #새로운 열(학년)을 모두 null로 채우기
df7.dropna(axis='columns', how='all')  #전부 null값인 행만 없애기
df7
```

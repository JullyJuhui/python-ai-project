```python
score = pd.read_csv('data/학생성적.csv')
no = f'{len(score)+1}번'  #str(len(score)+1) + '번'
no
```

```python
score.loc[len(score)]=[no, 0, 0, 0, 0, 0]
score
```

```python
filt = score['지원번호']=='8번'
idx = score[filt].index
if len(idx)==0: print('존재하지 않는 번호입니다.')
else:
    print(score[filt])
    score.drop(index=idx, inplace=True)
    print('삭제완료!')
score
```

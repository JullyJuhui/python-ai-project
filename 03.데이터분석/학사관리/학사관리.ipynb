```python
def pro_search(code):
    pro = pd.read_csv('교수.csv', index_col='교수번호')
    stu = pd.read_csv('학생.csv', index_col='학생번호')
    cou = pd.read_csv('강좌.csv', index_col='강좌번호')

    if code in pro.index:
        pro_row = pro.loc[code]
        print(code, pro_row['교수이름'], pro_row['교수학과'])

        print('담당학생-------------')
        filt = stu['지도교수']==code
        stu_rows = stu[filt]

        for idx, row in stu_rows.iterrows():
            print(idx, row['학생이름'], row['학생학과'])

        print('담당강좌-------------')
        filt = cou['담당교수']==code
        cou_rows = cou[filt]

        for idx, row in cou_rows.iterrows():
            print(idx, row['강좌이름'], row['강의시수'])
    
    else:
        print('해당교수가 없습니다.')

pro_search(221)  #이병렬 교수
```

```python
def stu_search(code):  #code = 학번
    stu = pd.read_csv('학생.csv', index_col='학생번호')
    pro = pd.read_csv('교수.csv', index_col='교수번호')
    enroll = pd.read_csv('수강.csv', index_col=['학생번호', '강좌번호'])
    enroll.fillna(0, inplace=True)
    cou = pd.read_csv('강좌.csv', index_col='강좌번호')

    stu_row = stu.loc[code]

    if not code in stu.index:
        print('해당 학생이 없습니다.')
        return

    advisor = stu_row['지도교수']
    pro_row = pro.loc[advisor]

    print(stu_row['학생이름'], stu_row['학생학과'], pro_row['교수이름'])
    print('\n수강과목')
    print('-'*30)
    enroll_rows = enroll.loc[code]

    sum = 0
    for idx, row in enroll_rows.iterrows():
        cou_row = cou.loc[idx]
        print(idx, cou_row['강좌이름'], row['신청일'], row['점수'])
        sum += row['점수']
    avg = sum/len(enroll_rows)
    print(f"평균: {avg:.2f}")

stu_search(92514009)
```

```python
#최대인원 
enroll = pd.read_csv('수강.csv')
group = enroll.groupby('강좌번호').size().to_frame('수강인원')
group.reset_index(inplace=True)
group
```

```python
merge = pd.merge(cou, group, how='left', on='강좌번호')
merge.fillna({'수강인원':0}, inplace=True)

merge['수강인원'] = merge['수강인원'].astype(int)
merge
```

```python
#수강신청한 강좌 수
def stu_list():
    stu = pd.read_csv('학생.csv')
    enroll = pd.read_csv('수강.csv')
    group = enroll.groupby('학생번호').size().to_frame('신청강좌 수')
    group.reset_index(inplace=True)

    merge = pd.merge(stu, group, how='left', on='학생번호')
    merge.fillna(0, inplace=True)
    merge['신청강좌 수'] = merge['신청강좌 수'].astype(int)
    merge.set_index('학생번호', inplace=True)
    return merge

stu_list()
```
